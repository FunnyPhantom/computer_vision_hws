```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'r')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('plot')
plt.show
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y,'b')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('OOP')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y, 'b')
axes1.set_xlabel('X1')
axes1.set_ylabel('Y1')
axes1.set_title('Axes 1')

axes2.plot(y,x,'r')
axes2.set_xlabel('X2')
axes2.set_ylabel('Y2')
axes2.set_title('Axes 2')
```

```python
fig, axes = plt.subplots()
print(fig)
print(axes)
axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
print(fig)
print(axes)
```

```python
for ax in axes:
    ax.plot(x,y,'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('plot')
fig
```

```python
fig, axes = plt.subplots(figsize=(3,3))
axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
fig.savefig("filename.png")
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,'b--',label='x^2')
ax.plot(x,x**3,'r',label='x^3')
ax.legend(loc=0)
```

```python
fig, axes = plt.subplots(1,3,figsize=(12,4))
axes[0].plot(x,x**2,x,x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x,x**4)
axes[1].set_title("tight_axes")

axes[2].plot(x,x**2,x,x**3)
axes[2].set_ylim([0,60])
axes[2].set_xlim([2,5])
axes[2].set_title("custom_axes_range")
```
